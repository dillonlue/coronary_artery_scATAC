```python
import pandas as pd
```

```python
df = pd.read_csv("/project/cphg-millerlab/CAD_QTL/coronary_QTL/epigenome/scATAC/deepsea/CAD_abs_diffs.tsv", sep='\t')
```

```python
snps = pd.read_csv("/scratch/dyl3pc/BPNet/external_resources/CAD.tsv", sep='\t', names=['chr', 'pos', 'rsid', 'ref', 'alt'])
```

```python
snps.index = snps.apply(lambda e: f"{e['chr']}_{e['pos']}", axis=1)
```

```python
chr_pos_snp_dict = snps.to_dict()['rsid']
```

```python
df['name'] = df.apply(lambda e: chr_pos_snp_dict[f"{e['chrom']}_{e['pos']}"], axis=1)
```

```python
df.to_csv("/project/cphg-millerlab/CAD_QTL/coronary_QTL/epigenome/scATAC/deepsea/CAD_abs_diffs.tsv", sep='\t', index=False)
```
